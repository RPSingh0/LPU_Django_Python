# Q3. Write a program to check if a number is prime or not. Example: 7 ==> True, 6 ==> False

## Taking n = 7 , output --> True
## Taking n = 6 , output --> False

```python
n = int(input("Enter a number:"))
factors = 0
flag = 0
for i in range(1,n+1):
    if n%i == 0:
        factors +=1
    if factors == 3:
        flag = 1
        break
if flag:
    print("False")
else:
    print("True")
```

# Q5. Write a program that asks the user for a number n and gives them the possibility to choose between computing the sum and computing the product of 1,…,n.

## Taking n = 5
## choice = sum , output --> Sum(1+2+3+....+5) = 15

## Taking n = 7
## choice = product , output --> Product(1.2.3...7) = 5040

```python
def fact(number):
    factorial = 1
    while(number != 1):
        factorial = factorial*number
        number -=1
    return factorial
def calculate(num, choice):
    if choice == "sum":
        print(f"Sum(1+2+3+....+{num}) = {sum(range(1,n+1))}")
    else:
        print(f"Product(1.2.3...{num}) = {fact(num)}")


n = int(input("Enter a number:"))
choice = input(f"Enter sum(1+2+3+....+{n}) or product(1.2.3...{n}) : sum/product:")
calculate(n, choice)
```

# Q9. Write a program which can compute the factorial of a given number.

## Taking n = 5, output --> Factorial of 5 is 120

```python
def fact(number):
    factorial = 1
    while(number != 1):
        factorial = factorial*number
        number -=1
    return factorial

n = int(input("Enter a number:"))
print(f"Factorial of {n} is {fact(n)}")
```

# Q13. Write a program that accepts a sentence (string) and calculate the number of letters and digits. Example: ‘this is a test sentence number 389’ ==> letters = 25 and digits = 3.

```python
sentence = input("Enter a sentence:")
letters = 0
digits = 0
for i in sentence:
    if i.isdigit():
        digits+=1
    elif i.isalpha():
        letters+=1
print(f"Letters = {letters} and Digits = {digits}")
```

# Q15. Write a program that accepts a sentence and calculate the number of upper case letters and lower case letters.

```python
sentence = input("Enter a sentence:")
upper_case = 0
lower_case = 0
for i in sentence:
    if i.isupper():
        upper_case+=1
    elif i.islower():
        lower_case+=1
print(f"Uppercase = {upper_case} and Lowercase = {lower_case}")
```

# 15. Make a password validator with the following checks. A website requires the users to input username and password to register. Write a program to check the validity of password input by users.
    Following are the criteria for checking the password:

1. At least 1 letter between [a-z]
     
2. At least 1 number between [0-9]
     
3. At least 1 letter between [@$#%&]

# Method 1
## Using regex

## Taking n = hejd$%367 , output --> Valid
## Taking n = jjeoa7736 , output --> Invalid

```python
import re 
pattern = re.compile(r"^(?=.+[a-z])(?=.+\d)(?=.+[@$#%&])[a-z\d@$#%&]{3,}$")
n = input("Enter a password to validate:")
if re.fullmatch(pattern, n):
    print("Valid")
else:
    print("Invalid")
```

# Method 2

## Taking n = hejd$%367 , output --> Valid
## Taking n = jjeoa7736 , output --> Invalid

```python
import string
a_z = string.ascii_lowercase
digits = list(map(str,range(0,10)))
char = list("@$#%&")
password = list(input("Enter you password to validate:"))
if any(x in a_z for x in password):
    if any(x in digits for x in password):
        if any(x in char for x in password):
            print("Valid")
        else:
            print("Invalid")
    else:
        print("Invalid")
else:
    print("Invalid")
```
